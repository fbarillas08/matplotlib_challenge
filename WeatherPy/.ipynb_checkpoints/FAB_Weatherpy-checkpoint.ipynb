```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
import requests
import json
import time
import csv
import os

from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Output File (CSV)
output_data_file = "output_data/cities.csv"
```

## Generate Cities List

```python
# world cities dataset selected from https://simplemaps.com/data/world-cities.
# This world cities list containst over 26,000 entries as of November 2020

worldcities_df = pd.read_csv('Resources/worldcities.csv')
worldcities_df
```

```python
# Cleanup by eliminating unnecessary columns
worldcities_df.drop(['city_ascii','iso2','iso3','admin_name','capital','population','id'], axis=1, inplace=True)
new_c_order = ['city','country','lat','lng']
worldcities_df[new_c_order]
```

```python
# List for holding Norther Hemisphere and Southern Hemisphere Cities
North_cities = []
South_cities = []
```

```python
North_cities = worldcities_df[worldcities_df.lat >=0]
len(North_cities)
```

```python
South_cities = worldcities_df[worldcities_df.lat <0]
len(South_cities)
```

```python
# Randomize Sample of Northern Hemisphere Cities

random_North_cities = North_cities.sample(350)
random_North_cities[new_c_order]
```

```python
# Randomize Sample of Southern Hemisphere Cities
random_South_cities = South_cities.sample(350)
random_South_cities[new_c_order]
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Config information for query
url = "http://api.openweathermap.org/data/2.5/weather?"
```

```python
for idx, row in random_North_cities.iterrows():
    city = row['city']
    query_url = url + "appid=" + weather_api_key + "&q=" + city
    weather_response = requests.get(query_url)
    weather_json = weather_response.json() 
    try:
        random_North_cities.loc[idx, 'Temperature'] = weather_json['main']['temp']
        random_North_cities.loc[idx, 'Humidity_pct'] = weather_json['main']['humidity']
        random_North_cities.loc[idx, 'Cloudiness_pct'] = weather_json['clouds']['all']
        random_North_cities.loc[idx, 'Windspeed'] = weather_json['wind']['speed']
    except KeyError:
        print(f"Not Found")
```

```python
random_North_cities
```

```python
# Drop NaN 
random_North_cities.dropna(inplace = True)
random_North_cities
```

```python
random_North_cities.describe()
```

```python
# Finding Weather Data for Southern Hemisphere Cities

for idx, row in random_South_cities.iterrows():
    city = row['city']
    query_url = url + "appid=" + weather_api_key + "&q=" + city
    weather_response = requests.get(query_url)
    weather_json = weather_response.json() 
    try:
        random_South_cities.loc[idx, 'Temperature'] = weather_json['main']['temp']
        random_South_cities.loc[idx, 'Humidity_pct'] = weather_json['main']['humidity']
        random_South_cities.loc[idx, 'Cloudiness_pct'] = weather_json['clouds']['all']
        random_South_cities.loc[idx, 'Windspeed'] = weather_json['wind']['speed']
    except KeyError:
        print(f"Not Found")
```

```python
# Drop NaN 
random_South_cities.dropna(inplace = True)
random_South_cities
```

```python
random_South_cities.describe()
```

# Merge Northern Hemisphere Dataframe with Southern Hemisphere Dataframe
### Export to csv
### Display dataframe statistics



```python
all_cities_df = random_North_cities.append(random_South_cities)
#sort cities based on latitude
all_cities_df = all_cities_df.sort_values('lat', ascending = False)
all_cities_df
```

```python
all_cities_df.describe()
```

# Cities with Humidity 100% or Greater


```python
super_humid_cities_df = all_cities_df[all_cities_df.Humidity_pct >= 100]
super_humid_cities_df
```

# Eliminate Super Humid Cities from the Main Dataframe

```python
all_cities_df = all_cities_df[all_cities_df.Humidity_pct < 100]
```

```python
# Visualize Distribution of Humidity using a Box and Whisker Plot
cities_humidities = all_cities_df['Humidity_pct']
fig1, ax1 = plt.subplots()
ax1.set_title("% Humidity in Cities Around the World")
ax1.set_ylabel("% Humidity")
ax1.set_xlabel("All Cities Analysis")
ax1.boxplot(cities_humidities)
plt.show()
```

```python
# Make a new DataFrame equal to the city data to drop all humidity outliers by index.
# Passing "inplace=False" will make a copy of the city_data DataFrame, which we call "clean_city_data".

Humidity_quartiles = cities_humidities.quantile([.25,.5,.75])

Humidity_upperQ = Humidity_quartiles[0.75]
Humidity_median = Humidity_quartiles[0.5]
Humidity_lowerQ = Humidity_quartiles[0.25]
Humidity_IQR = Humidity_upperQ - Humidity_lowerQ

Humi_upperbound = Humidity_upperQ + 1.5*Humidity_IQR
Humi_lowerbound = Humidity_lowerQ - 1.5*Humidity_IQR


print(f"The upper quartile of Humidity is: {Humidity_upperQ}")
print(f"The the median of Humidity is:     {Humidity_median} ")
print(f"The lower quartile of Humidity is: {Humidity_lowerQ}")
print(f"The IQR of Humidity is:            {Humidity_IQR}")
print(f"Outliers Upper Bound is:           {Humi_upperbound}")
print(f"Outliers Lower Bound is:           {Humi_lowerbound}")
```

# Eliminating the Outliers

```python
clean_city_data_df=all_cities_df[all_cities_df.Humidity_pct > Humi_lowerbound]
clean_city_data_df
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = random_North_cities['lat']
y_values = random_North_cities['Temperature']

(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))

plt.scatter(x_values, y_values, edgecolor = 'red', c=y_values ,s=50, cmap ='Blues')
plt.title('Northern Hemisphere Latitude vs Temperature Scatter Plot \n Date of Analysis: 02.04.21', fontdict ={"fontsize":20})
plt.xlabel('Latitude', fontdict ={"fontsize":18})
plt.ylabel('Temperature (K)',fontdict ={"fontsize":18})
plt.grid(color ='k', alpha =0.15)
plt.tight_layout(pad =0.8, w_pad = 1.4, h_pad=1.0)
plt.rcParams['figure.figsize']=(16,12)

plt.plot(x_values,regress_values,"k-")

cbar = plt.colorbar()
cbar.set_label('Temperature in K')

print(f"The R-value is: {rvalue**2}")
print(line_eq)
print(f"Observation:")
print(f"As the latitude decreases so does the temperature in the Northern Hemisphere, the linear regression model is adequate for this data")
plt.savefig('output_data/NorthCities_Temp_vs_Lat.png')
plt.show()
```
